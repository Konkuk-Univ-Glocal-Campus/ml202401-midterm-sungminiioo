```python
import seaborn as sns

# 데이터 불러오기
data = pd.read_csv("data.csv", encoding='utf-8')

# '상세영업상태명'이 '영업중'인 행 선택
data = data[data['상세영업상태명'] == '영업중']

# '도로명전체주소' 열에서 지역 정보 추출
data['지역'] = data['도로명전체주소'].str.split(' ', expand=True).iloc[:, 0]

# 전체 지역 중 가장 많이 등장한 상위 5개 지역 선택
top_locations = data['지역'].value_counts().head(10)

# 시각화
plt.figure(figsize=(10, 7))
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45)



plt.title('Top 10 Regions with Operating Stores Nationwide', color='blue')
plt.xlabel('Region')
plt.ylabel('Number of Stores')
plt.show()
```

```python
#데이터 선별
data = data[["좌표정보(x)", "좌표정보(y)"]]
```

```python
#결측치 제거
data = data.dropna()
```

```python
# Heatmap 추가
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#raw data 시각화

plt.scatter(data.iloc[:, 0], data.iloc[:, 1], cmap="gist_rainbow")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#산점도와 커널 밀도 추정
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data.iloc[:, 0], y=data.iloc[:, 1], cmap="gist_rainbow", s=50)
sns.kdeplot(x=data.iloc[:, 0], y=data.iloc[:, 1], cmap="coolwarm", levels=5, linewidths=1.5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot with Kernel Density Estimation")
plt.show()
```

```python
#K-means 생성 및 학습
#n_clusters를 처음에 9로 정한 이유: 전국 8도 + 울릉도
km = KMeans(n_clusters=9, n_init=10, random_state=20)

km.fit(data)
```

```python
labels = km.labels_

plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels, cmap='jet',
edgecolor='k', s=10)
```

```python
#Elblow Method를 이용한 타당성 평가
%pip install yellowbrick
from yellowbrick.cluster import KElbowVisualizer

k = 0
km = KMeans(n_clusters=k, n_init=10, random_state=20)
visualizer = KElbowVisualizer(km, k=(1, 20), timings=False)
visualizer.fit(data)
visualizer.show()
```

```python
#새로운 K-Means 생성 및 학습
new_km = KMeans(n_clusters=4, n_init=10, random_state=20)

new_km.fit(data)
```

```python
new_labels = new_km.labels_

plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=10)
```
