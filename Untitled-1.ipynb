```python
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 준비 및 전처리
data = pd.read_csv("data.csv", encoding='utf-8')

data.head()
```

```python
# 시각화를 통해 결측치 확인
plt.figure(figsize=(10, 6))
plt.title('Missing Data Visualization')
plt.imshow(data.isnull(), aspect='auto', cmap='viridis')
plt.colorbar(label='Missing Values')
plt.show()
```

```python
# '문화체육업종명'이 '관광펜션업'인 행 선택
data = data[data['문화체육업종명'] == '관광펜션업']

# '도로명전체주소' 열에서 지역 정보 추출
data['지역'] = data['도로명전체주소'].str.split(' ', expand=True).iloc[:, 0]

# 전체 지역 중 가장 많이 등장한 상위 10개 지역 선택
top_locations = data['지역'].value_counts().head(10)

# 시각화
plt.figure(figsize=(10, 7))
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45)
plt.title('Top 10 Regions with the Most Tourist Pensions Nationwide', color='blue')
plt.xlabel('Region')
plt.ylabel('Number of Tourist Pensions')
plt.show()
```

```python
# '문화체육업종명'이 '관광펜션업'이면서 '상세영업상태명'이 '폐업' 또는 '휴업'인 행 선택
closed_data = data[(data['문화체육업종명'] == '관광펜션업') & ((data['상세영업상태명'] == '폐업') | (data['상세영업상태명'] == '휴업'))]

# '도로명전체주소' 열에서 지역 정보 추출
closed_data['지역'] = closed_data['도로명전체주소'].str.split(' ', expand=True).iloc[:, 0]

# 전체 지역 중 가장 많이 등장한 상위 10개 지역 선택
top_closed_locations = closed_data['지역'].value_counts().head(10)

# 시각화 - 폐업 및 휴업중인 관광펜션업의 지역별 분포
plt.figure(figsize=(10, 7))
sns.barplot(x=top_closed_locations.index, y=top_closed_locations.values)
plt.xticks(rotation=45)
plt.title('Top 10 Regions with the Most Closed or Temporarily Closed Tourist Pensions Nationwide', color='red')
plt.xlabel('Region')
plt.ylabel('Number of Tourist Pensions (Closed or Temporarily Closed)')
plt.show()
```

```python
# '상세영업상태명'이 '영업중'인 행 선택
open_data = data[(data['문화체육업종명'] == '관광펜션업') & (data['상세영업상태명'] == '영업중')]

# '도로명전체주소' 열에서 지역 정보 추출
open_data['지역'] = open_data['도로명전체주소'].str.split(' ', expand=True).iloc[:, 0]

# 전체 지역 중 가장 많이 등장한 상위 10개 지역 선택
top_open_locations = open_data['지역'].value_counts().head(10)

# 시각화 - 영업중인 관광펜션업의 지역별 분포
plt.figure(figsize=(10, 7))
sns.barplot(x=top_open_locations.index, y=top_open_locations.values)
plt.xticks(rotation=45)
plt.title('Top 10 Regions with the Most Operating Tourist Pensions Nationwide', color='blue')
plt.xlabel('Region')
plt.ylabel('Number of Tourist Pensions (Operating)')
plt.show()
```

```python
# '문화체육업종명'이 '관광펜션업'인 행 선택
tourist_pensions = data[data['문화체육업종명'] == '관광펜션업']

# '도로명전체주소' 열에서 지역 정보 추출
tourist_pensions['지역'] = tourist_pensions['도로명전체주소'].str.split(' ', expand=True).iloc[:, 0]

# Pivot table을 이용하여 지역별 관광펜션업의 수 계산
pension_counts = tourist_pensions.pivot_table(index='지역', aggfunc='size').reset_index(name='펜션 수')

# Heatmap 그리기
plt.figure(figsize=(12, 8))
sns.heatmap(pension_counts.pivot_table(index='지역', columns='펜션 수', aggfunc=len), cmap='YlGnBu', linewidths=0.5, linecolor='gray')
plt.title('Distribution of Tourist Pensions by Region', fontsize=16)
plt.xlabel('Number of Tourist Pensions', fontsize=14)
plt.ylabel('Region', fontsize=14)
plt.show()
```

```python
# '문화체육업종명'이 '관광펜션업'인 행 선택
tourist_pensions = data[data['문화체육업종명'] == '관광펜션업']

# 선택한 열 목록
selected_columns = ['건물용도명', '총층수',  '지상층수','시설규모']

# 선택한 열들에 대한 상자 그림 그리기
for col in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=tourist_pensions[col], orient="v")
    plt.title(f'Boxplot of {col}')
    plt.ylabel('Values')
    plt.show()
```

```python
# K-means 생성 및 학습
# n_clusters를 처음에 9로 정한 이유: 전국 8도 + 울릉도
km = KMeans(n_clusters=9, n_init=10, random_state=20)

# '좌표정보(x)'와 '좌표정보(y)' 열 선택
coordinates = tourist_pensions[["좌표정보(x)", "좌표정보(y)"]]

# 결측치 제거
coordinates = coordinates.dropna()

# K-means 모델에 데이터 학습
km.fit(coordinates)
```

```python
# 학습된 클러스터링 결과를 이용하여 산점도 그리기
labels = km.labels_

plt.figure(figsize=(10, 6))
plt.scatter(coordinates.iloc[:, 0], coordinates.iloc[:, 1], c=labels, cmap='jet',
            edgecolor='k', s=10)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("K-means Clustering")
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

# KElbowVisualizer를 사용하여 최적의 클러스터 개수 결정
k_range = range(1, 20)
visualizer = KElbowVisualizer(KMeans(n_init=10, random_state=20), k=k_range, timings=False)

# 데이터에 대해 Elbow Method 적용
visualizer.fit(coordinates)
visualizer.show()
```

```python
# 새로운 K-means 모델 생성 및 학습
new_km = KMeans(n_clusters=4, n_init=10, random_state=20)
new_km.fit(coordinates)
```

```python
# 학습된 새로운 클러스터링 결과를 이용하여 산점도 그리기
new_labels = new_km.labels_

plt.figure(figsize=(10, 6))
plt.scatter(coordinates.iloc[:, 0], coordinates.iloc[:, 1], c=new_labels, cmap='jet',
            edgecolor='k', s=10)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("New K-means Clustering")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# K-means 모델의 예측 결과
labels = km.labels_

# 실루엣 점수 계산
silhouette_avg = silhouette_score(coordinates, labels)

print(f"The average silhouette score of the K-means clustering is: {silhouette_avg}")
```
